```python
import pandas as pd
import matplotlib
import numpy as np
import seaborn as sb
```

```python
df = pd.read_csv('gd_tweets_2023_FULL.csv')
```

```python
##Dataframe overview
```

```python
df
```

```python
##Simple bar graph that shows top 10 brands and the amount of tweets they had
ax = df.groupby(['brand_ad_name'])['brand_ad_name'].count().nlargest(10).plot(kind='bar')

ax.set_xlabel=('Brand Name')
ax.set_ylabel=('Number of Tweets')
ax.set_title('Top 10 brands with the most tweets')
```

```python
df2 = df.groupby(['brand_ad_name'])['brand_ad_name'].count().nlargest(30)
df2
```

```python
df3 = df.groupby(['keywords'])['keywords'].count().nlargest(30)
df3
```

```python
df4 = df.groupby(['time_period_of_the_game'])['keywords'].count().nlargest(10)
df4
```

```python
##Lineplot to show the amount of tweets throughout the game
df4.plot(kind='line', y='time_period_of_the_game', figsize=(10,10))
```
